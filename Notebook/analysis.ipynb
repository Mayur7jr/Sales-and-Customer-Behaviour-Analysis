```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r"E:\Data Analyst Bootcamp\Python Project\data\advanced_sales_dataset_2021_2024.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
mode_value=data[data['Category'].notna()]['Category'].mode()[0]
print(mode_value)
```

```python
data['Filled_Category']=data['Category'].fillna(mode_value)
```

```python
data.head()
```

```python
data.drop('Category',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data["Price"].dtype
```

```python
data['Filld_Price']=data['Price'].fillna(data['Price'].mean())
```

```python
data.drop("Price",axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
mode_value1=data[data['Payment_Method'].notna()]['Payment_Method'].mode()[0]
print(mode_value1)
```

```python
data['Filled_Payment_Method']=data['Payment_Method'].fillna(mode_value1)
```

```python
data.drop("Payment_Method",axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
city_mode=data[data['City'].notna()]['City'].mode()[0]
print(city_mode)
```

```python
data['Filled_City']=data['City'].fillna(city_mode)
```

```python
data.drop('City',axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.duplicated().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x='Filled_Category',y='Total_Amount',data=data)
plt.title("Total Sales")
plt.show()
```

```python
sns.boxplot(x='Filled_Category',y='Filld_Price',data=data)
```

```python
data['Order_Date']=pd.to_datetime(data['Order_Date'])

sales_by_year=data.groupby(data['Order_Date'].dt.year)["Total_Amount"].sum()

print(sales_by_year)
```

```python
plt.figure(figsize=(5,5))
plt.plot(sales_by_year)
plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Total Amount")
plt.show()
```
